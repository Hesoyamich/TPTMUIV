```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df.replace(True, 1, inplace=True)
df.replace(False, 0, inplace=True)
df
```

```python
plt.plot(df["perlin_value"], df['selected_tile'], 'bo')
plt.xlabel("Шум перлина")
plt.ylabel("Номер клетки")
plt.show()
```

```python

mean_tiles = df.groupby("selected_tile")["possible_tiles"].mean()
plt.bar(mean_tiles.index, mean_tiles.values)
plt.xlabel("Номер клетки")
plt.ylabel("Среднее количество возможных вариантов")
plt.show()
```
